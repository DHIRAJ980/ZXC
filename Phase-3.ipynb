```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array ([
    [1, 320000, 389000, 120000, 330000],
    [2, 630000, 333728, 923718, 366781],
    [3, 849200, 821030, 325188, 381003],
    [4, 327389, 390284, 298840, 472894],
    [5, 493902, 480224, 409238, 380402],
])

print("=== Zomato sales analysys ===")
print("\n Sales data shape", sales_data.shape)
print("\n *** Sample data for 1st 3 restaurant *** \n ", sales_data [: 3])
print("\n *** Sample data for 1st 5 restaurant *** \n ", sales_data [:, 1:])
```

```python
# Totel sales per year
print(np.sum(sales_data[:,1:], axis=1))
yeary_totel = np.sum(sales_data[:, 1:], axis =0)
print(yeary_totel)

# Minimum sales per restaurant
min_sales =np.min(sales_data[:, 1:], axis=1)
print(min_sales)

#maximum sales per year
max_sales =np.max(sales_data[:, 1:], axis =0)
print(max_sales)

# Average sales per restautant
avg_sales = np.mean(sales_data[:, 1:], axis=0)
print(avg_sales)
```

```python
cumsum = np.cumsum(sales_data[:, 1:], axis =1)
print(cumsum)

plt.figure(figsize=(8,6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average cumulative sales accross all restaurant ")
plt.xlabel("years")
plt.ylabel("salea")
plt.grid(True)
plt.show
```

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([6, 7, 8, 9, 10])

print("Vector addition: ", vector1 + vector2)

print("\n Multiplication:", vector1 * vector2)

print("\n Dot Product", np.dot(vector1, vector2))

angle = np.arccos(np.dot(vector1, vector2)/ (np.linalg.norm(vector1)
                  *np.linalg.norm(vector2)))
print(angle)
```
